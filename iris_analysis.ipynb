```python
import pandas as pd
data = pd.read_csv('iris_data.csv')

import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame(data)
print(df)

# Display first few rows
print("First 5 Rows:")
print(df.head())

# Check structure
print("\nDataset Info:")
print(df.info())

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Task 2: Basic Data Analysis
print("\nDescriptive Statistics:")
print(df.describe())

# Group by class and calculate mean
print("\nGrouped Means:")
print(df.groupby("class").mean(numeric_only=True))

# Task 3: Data Visualizations
sns.set(style="whitegrid")

# Line Plot (Petal Length over Index)
plt.figure(figsize=(8, 4))
plt.plot(df.index, df["petal length"], marker='o')
plt.title("Petal Length Over Observations")
plt.xlabel("Index")
plt.ylabel("Petal Length")
plt.grid(True)
plt.show()

# Bar Chart (Average Petal Length per Class)
plt.figure(figsize=(6, 4))
sns.barplot(x="class", y="petal length", data=df, errorbar=None)
plt.title("Average Petal Length per Class")
plt.ylabel("Petal Length")
plt.show()

# Histogram (Sepal Width Distribution)
plt.figure(figsize=(6, 4))
sns.histplot(df["sepal width"], bins=5, kde=True)
plt.title("Distribution of Sepal Width")
plt.xlabel("Sepal Width")
plt.show()

# Scatter Plot (Sepal Length vs Petal Length)
plt.figure(figsize=(6, 4))
sns.scatterplot(x="sepal length", y="petal length", hue="class", data=df)
plt.title("Sepal Length vs. Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.legend()
plt.show()
```
